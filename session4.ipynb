```python
# .agg() allows you to apply a function or a list of function names to be executed along one of the axis of the DataFrame, default 0
def pct40(column):
    return column.quantile(0.4)
def iqr(column): # A custom IQR function
    return column.quantile(0.75) - column.quantile(0.25)
    
print(homelessness["family_members"].agg(["quantile",sum,"max",iqr]) )
```
